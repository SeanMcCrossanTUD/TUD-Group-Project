# Create dirty dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ks-projects-201801.csv')
```

```python
sampled_df = df.sample(n=60000, random_state=1)
```

```python
sampled_df.isnull()
```

```python
import pandas as pd
import numpy as np

def dirty_dataset(df):
    # Set random seed for reproducibility
    np.random.seed(0)

    # Step 1: Inserting Null Values
    # Choose 1% of the total entries in the DataFrame and set them to NaN
    num_cells = df.size
    num_null = int(num_cells * 0.01)
    for _ in range(num_null):
        row_idx = np.random.randint(0, df.shape[0])
        col_idx = np.random.randint(0, df.shape[1])
        df.iat[row_idx, col_idx] = np.nan

    # Step 2: Deleting Rows
    # Randomly choose 2 rows to delete
    drop_indices = np.random.choice(df.index, 2, replace=False)
    df = df.drop(drop_indices)

    # Step 3: Duplicating Rows
    # Randomly choose 2 rows to duplicate
    duplicate_indices = np.random.choice(df.index, 2, replace=False)
    duplicate_rows = df.loc[duplicate_indices]
    df = pd.concat([df, duplicate_rows])

    # Reset the index
    df = df.reset_index(drop=True)

    return df

# Use the function
dirty_df = dirty_dataset(sampled_df)
```

```python
duplicate_rows = dirty_df[dirty_df.duplicated(keep=False)]
duplicate_rows
```

```python
def check_dataset_errors(df):
    # Check for duplicate rows
    duplicate_rows = df[df.duplicated(keep=False)]

    # Check for null values
    null_values = df.isnull().sum().sum()

    # Check the DataFrame size
    num_rows, num_columns = df.shape

    print(f"Number of duplicate rows: {len(duplicate_rows)}")
    print(f"Number of null values: {null_values}")
    print(f"Number of rows: {num_rows}")
    print(f"Number of columns: {num_columns}")

# Use the function
check_dataset_errors(dirty_df)
```

```python
# Saving the dirty DataFrame as a CSV file
dirty_df.to_csv('dirty_startup_dataset.csv', index=False)
```

# Clean this dataset

```python
def clean_dataset(df):
    # Removing duplicate rows
    df = df.drop_duplicates()

    # Option 1: Removing rows with null values
    # df = df.dropna()

    # Option 2: Filling null values with 0
    # df = df.fillna(0)

    # Option 3: Filling null values with the mean of the column
    df = df.fillna(df.mean())

    return df

# Use the function
cleaned_df = clean_dataset(dirty_df)
```
